__Task 3:__ Calculate the mean, max and min value of the age and the hours-per-week columns. Please note, for the purpose of this exercise, you are required to write your own calculation instead of using helper functions. 

Hint: In order to do this calcualtion without helper methods, you need to use loop to enumerate the values.

```python
def findMean(data):
    total = 0
    #add each value to the total
    for i in data:
        total += i
    #divide the total by the length
    mean = total / len(data)
    return mean

def findMax(data):
    max = 0
    #loop through
    for i in data:
        #if current value is more than max, set max as current value
        if (i > max):
            max = i
    return max

def findMin(data):
    min = 9999
    #loop through
    for i in data:
        #if current value is less than min, set min as current value
        if (i < min):
            min = i
    return min



hrsData = adult_data['hours-per-week']
ageData = adult_data['age']

#print max, min and mean of hrs-per-week
print('Max Hours per week: ' + str(findMax(hrsData)))
print('Min Hours per week: ' + str(findMin(hrsData)))
print('Mean Hours per week: ' + str(findMean(hrsData)))
    
#print max, min and mean of age
print('Max Age: ' + str(findMax(ageData)))
print('Min Age: ' + str(findMin(ageData)))
print('Mean Age: ' + str(findMean(ageData)))
```

__Task 5:__ Find out from the data, the percentage of male with salary > 50K and the percentage of female with salary > 50K.

```python
#put all male data into df
malepop = adult_data[(adult_data['sex'] == ' Male')]
#put all female data into df
femalepop = adult_data[(adult_data['sex'] == ' Female')]
#put all male data where salary > 50k into df 
malepopSalary50k = malepop[(malepop['salary'] == " >50K")]
#put all female data where salary >50k into df
femalepopSalary50k = femalepop[(femalepop['salary'] == " >50K")]

#find percentages by using length of dfs
percMale50k = (len(malepopSalary50k) / len(malepop)) * 100
percFemale50k = (len(femalepopSalary50k) / len(femalepop)) * 100

#print
print("Percentage of males earning >50K is " + str(percMale50k) + "%")
print("Percentage of females earning >50K is " + str(percFemale50k) + "%")
```

__Bonus Task:__ Calculate the mode, median and variance of age and hours-per-week attributes. Note: the requirement here is similar to Task 3, you have to write your own.

```python
import numpy


def calcMode(data) :
    maxcount = 0
    mode = 0
    #array used for count of each element
    count = numpy.zeros(shape=(len(data),1))
    #loop through each item
    for i in data:
        #increase count of element
        count[i] += 1;
        #if maxcount is less than current item count, set maxcount to current item count and set mode to current item
        if (maxcount < count[i]):
            maxcount = count[i]
            mode = i
    return mode

#get data
age_data = adult_data['age']
hrs_data = adult_data['hours-per-week']
#call mode method on age and hrs per week and print
print("Age mode is " + str(calcMode(age_data)))
print("Hrs-per-week mode is " + str(calcMode(hrs_data)))


```

```python
def calcMedian(data):
    #sort
    sortedData = sorted(data)
  
    #if length is odd
    if len(sortedData) % 2 == 1:
        return sortedData[(len(sortedData)+1) // 2-1]
    #otherwise 
    else:
        below = sortedData[len(sortedData) // 2-1]
        above = sortedData[len(sortedData)//2]
        return (float(below + above)) // 2 

#get data
age_data = adult_data['age']
hrs_data = adult_data['hours-per-week']

#call median on age and hrs per week and print
print("Age median is: " + str(calcMedian(age_data)))
print("Hrs-per-week median is: " + str(calcMedian(hrs_data)))
```

```python
def calcVar(data):
    total = 0
    variance = 0
    #using mean function from before
    mean = findMean(data)
    for i in data:
        #subtract mean
        a = (i - mean)
        #square
        squared = a * a
        #add to toal
        total += squared
    #divide total by lenght of data (-1). This gives the variance
    variance = (total / (len(data) - 1)) 
    return variance

#get data
age_data = adult_data['age']
hrs_data = adult_data['hours-per-week']

#call variance on age and hrs per week
print("Age variance is " + str(calcVar(age_data)))
print("Hrs-per-week variance is " + str(calcVar(hrs_data)))
```
